# Web_Scrapping

######     Web scraping is a computer software technique of extracting information from websites.

```python
#import the library used to query a website

from urllib.request import urlopen
```

```python
#specifying the url

page=urlopen("https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India")
```

Beautiful Soup provides a few simple methods and Pythonic idioms for navigating,
searching, and modifying a parse tree: a toolkit for dissecting a document and
extracting what you need. It doesn't take much code to write an application.

Beautiful Soup automatically converts incoming documents to Unicode and outgoing
documents to UTF-8. You don't have to think about encodings, unless the document
doesn't specify an encoding and Beautiful Soup can't autodetect one.

```python
#import the beautifulsoup function to parsed the data returned from the website

from bs4 import BeautifulSoup
```

```python
#Parse the html in the 'Page' variable and store in Beautiful Soup format

soup=BeautifulSoup(page)
all_tables=soup.find_all('table')
right_table=soup.find('table',class_="wikitable sortable plainrowheaders")
```

```python
#Generate Lists
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]
for row in right_table.findAll("tr"):
    cells=row.findAll("td")
    #To store 2nd column data
    states=row.findAll("th")
    #Only extract the table body not heading
    if len(cells)==6:
        A.append(cells[0].find(text=True))
        B.append(states[0].find(text=True))
        C.append(cells[1].find(text=True))
        D.append(cells[2].find(text=True))
        E.append(cells[3].find(text=True))
        F.append(cells[4].find(text=True))
        G.append(cells[5].find(text=True))
        
        
```

```python
#import pandas to convert list to data frame
import pandas as pd
df=pd.DataFrame(A,columns=["Number"])
df['States']=B
df['Admin_capital']=C
df['legistative_capital']=D
df['Judiciary_capital']=E
df['Year_capital']=F
df['Formal_capital']=G
```

```python
#print the table
df
```

```python
#import the library used to query a website
from urllib.request import urlopen



#specifying the url
page=urlopen("https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_area")
```

```python
#import the beautifulsoup function to parsed the data returned from the website

from bs4 import BeautifulSoup 
soup=BeautifulSoup(page) 
all_tables=soup.find_all('table') 
right_table=soup.find('table',class_="wikitable sortable") 
```

```python
#Generate Lists

A=[] 
B=[] 
C=[] 
D=[] 
E=[] 
F=[] 
G=[] 
for row in right_table.findAll("tr"): 
    states=row.findAll("th") 
    cells=row.findAll("td") 
    if len(cells)==7: 
        A.append(cells[0].find(text=True)) 
        B.append(cells[1].find(text=True)) 
        C.append(cells[2].find(text=True)) 
        D.append(cells[3].find(text=True)) 
        E.append(cells[4].find(text=True)) 
        F.append(cells[5].find(text=True)) 
```
